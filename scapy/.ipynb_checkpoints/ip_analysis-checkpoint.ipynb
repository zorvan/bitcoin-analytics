```python
# https://en.bitcoin.it/wiki/Protocol_documentation
from scapy.all import *
from pprint import pprint

dic = {}
def pkt_callback(packet):
    global dic
    src = packet[IP].src
    if src in dic.keys():
        dic[src] += 1
    else:
        dic[src] = 1
```

```python
# from pprint import pprint

counter = 0
while(counter<10):
    p=sniff(count=100, filter="tcp port 8333", prn=pkt_callback)
    print(dic, end="\r")
    counter += 1
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.DataFrame(list(dic.items()))
df.sort_values(1, ascending=False)
```
